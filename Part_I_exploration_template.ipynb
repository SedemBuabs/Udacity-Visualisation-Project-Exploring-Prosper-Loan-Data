## Introduction

### What  are the Distribution of Categorical Variables 

```python
# Define a function to visualise Categorical Variables
def distribution_cat(column_name, rotation, df):
    """  
    
    Returns Visual showing the distribution of the specified variable in the dataframe. 
    Input: Takes a dataframe, a variable name, and a number(int).
    
    df: The dataframe
    column_name: The column's whose distribution you wish to view.
    rotation: This specifies the rotation of the tick labels.
    
    Output:
    Returns a visual showing distribution.
    
    """
    # Plot
    sb.countplot(data = df, x=column_name, color='Grey')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column_name}')
    plt.xticks(rotation = rotation)
    plt.show()
```

```python
def hist_plot(column_name, bins, df):
    """  
    
    Returns Visual(Histogram) showing the distribution of the specified variable in the dataframe. 
    Input: Takes a dataframe, a variable name, and a number(int).
    
    df: The dataframe
    column_name: The column's whose distribution you wish to visualise.
    bins: This specifies the number of bins of the Historgram.
    
    Output:
    Returns a visual showing a Histogram distribution.
    
    """
    plt.figure(figsize=[15,8])
    sb.histplot(df[column_name], bins=bins, color ='Grey')
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.title(f'Distribution of {column_name}')
    plt.show()
```

## Bivariate Exploration

> In this section, the aim is to uncover intriguing associations that demonstrate the impact of a variable on another, specifically focusing on the target variables: Loan Original Amount and Prosper Rating.

### What is the correlation between all numerical variables?

```python
# Create list of numerical variables of interest
num_variables =['BorrowerAPR',
 'BorrowerRate',
 'LenderYield',
 'ProsperRatingNumeric',
 'EmploymentStatus',
 'EmploymentStatusDuration',
 'IsBorrowerHomeowner',
 'BankcardUtilization',
 'AvailableBankcardCredit',
 'DebtToIncomeRatio',
 'IncomeVerifiable',
 'StatedMonthlyIncome',
 'LoanOriginalAmount',
 'Recommendations']

# Subset all numerical columns
loan_num = loan[num_variables]

# compute the correlation between variables
corr = loan_num.corr().round(2)

# hide upper diagonal of heatmap since it presents duplicate information
mask = np.triu(np.ones_like(corr,dtype=bool))

# Visualise the correlation between the numerical variables using a heatmap
plt.figure(figsize=(10,10))
sb.heatmap(corr,
            annot = True, mask = mask,
            cmap='coolwarm')
plt.title("Correlation Between Variables")
plt.show()
```

#### Observation

>BorrowerAPR has a strong positive correlation with Borrower Rate and Lender Yield.

>Prosper Rating has a strong negative correlation with Borrower APR, Borrower Rate and Lender Yield.

> The remaining variable pairs have no correlation with a few pairs like prosper rating and loan amount, loan amount and borrower APR being fairly moderately correlated.
